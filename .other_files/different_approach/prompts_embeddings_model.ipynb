```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np

prompt_file = "../prompts.xlsx"
df = pd.read_excel(prompt_file)

model = SentenceTransformer("BAAI/bge-large-en-v1.5")

embeddings = {}
for _, row in df.iterrows():
    effect = row.iloc[0]
    text = row.iloc[1]
    emb = model.encode(text, normalize_embeddings=True)
    embeddings.setdefault(effect, []).append(emb)

# Save embeddings for later
for effect, embs in embeddings.items():
    np.save(f"embeddings_{effect}.npy", np.array(embs))
```
